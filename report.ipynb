## Exploratory Data Analysis (EDA) ## 

The model is based on the movies, ratings and tags available in the mentioned data set ml-latest. All datasets contain different data for movies but the datasets are possible to link together by their movie id

Movies.csv contains movie id, title, and genres for individual movies. The title in the dataset includes both the movie title and its year of release. The genres for each movie are grouped together  but separated by a vertical bar [|]. The number of rows is 86537 and the number of valid genres are 19.   

ratings.csv contains movie ratings from individual users: movie id, user id, rating (0-5) and timestamp.

tags.csv contains individual tag for movies from individual users and timestamp.

Timestamp is the number of milliseconds that have passed since the beginning of time (1 jan 1970, Unix time). With timestamp taken into consideration for ratings, one could possibly make different suggestions for the user depending on what the time is. For example: Horror movies might be suggested during nighttime but penaltized during daytime. For this model, the timestamp has been discarded.

## Model ##

The recommendation system is based on a content-filtering approach, combining genres and user-generated tags into a textual representation for each movie. Collaborative filtering was evaluated but abandoned as the size of the matrix exceeded 22 MB RAM. 

The content filtering matrix is of shape (n_movies, n_features) where the last column consists of a list of combined genre and tag words for every movie. Cosine similarity calculates the simularity of every movie based on the combined genres and tags.

The code is split into 4 methods: load data files, build feature matrix, compute simularity and recommend. The purpuse of these fore methods is to make the code more readable.

load_data_files() imports movies, tags and ratings and saves them as dataframes.

build_feature_matrix() filters unpopular movies and groups genres together with tags into a list of words for each movie

compute_similarity() creates a simularity matrix containing all movies. TF-IDF is used to transform textual data into numerical feature vectors by weighting terms based on their frequency in a movie and their rarity across all movies.

recommend() makes a list of the top 5 similar movies based on the input movie, prints them into a table in the terminal and returns them as a dataframe
