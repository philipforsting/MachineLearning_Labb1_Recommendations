```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

movies_df = pd.read_csv('data/movies.csv')
movies_df
```

An obervation is that column genres holds strings separated with "|".

My thought is to extract all unique genres in order to create colums for them.

```python
movies_genres_df = movies_df["genres"]
movies_genres_list = []
movies_genres_set = set()

for genre_str in movies_genres_df:          
    genre_list = genre_str.split("|")       # Spliting the string of genres separated by "|"
    for genre in genre_list:
        movies_genres_set.add(genre)        # attempting to add every extracted genre to set. Set doews not allow duplicates so each element will be unique.

movies_genres_list = list(movies_genres_set)
movies_genres_list.sort()
movies_genres_list
```
